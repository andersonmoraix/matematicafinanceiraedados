# GUIA PRÁTICO: TUPLAS E ESTRUTURAS DE DADOS NO PYTHON

```python
# 1. CRIAÇÃO DE TUPLAS

# Tupla padrão (múltiplos elementos com parênteses)
coordenada_1 = (10, 20)

# Tupla sem parênteses (a vírgula é a estrutura principal)
coordenada_2 = 10, 20

# Tupla de um elemento único (Sintaxe correta exige a vírgula!)
elemento_unico = (15,)

# Sem a vírgula, o Python interpreta apenas como um inteiro comum
nao_e_tupla = (15)
```

```python
# 2. ACESSO AOS ELEMENTOS E IMUTABILIDADE

coordenada_geografica = (-30.0346, -51.2177)

# Acessando pelo índice numérico (começa sempre em 0)
print(f"Latitude: {coordenada_geografica[0]}")
print(f"Longitude: {coordenada_geografica[1]}")

# IMUTABILIDADE: Se tentarmos fazer a atribuição abaixo, gerará um erro:
# coordenada_geografica[0] = -31.0  # TypeError

```

```python
# 3. DESEMPACOTAMENTO (UNPACKING)

# Extraindo os valores internos diretamente para variáveis independentes
lat, long = coordenada_geografica
print(f"Desempacotamento -> Lat: {lat} | Long: {long}")
```

```python
# 4. LISTAS

# Ideal para coleções que podem mudar ao longo do tempo
tarefas = ["estudar", "treinar"]
tarefas.append("ler") # Modificando a lista: adicionando um novo elemento
print(f"Lista dinâmica: {tarefas}")
```

```python
# 5. SETS (Conjuntos)

# Não possuem ordem fixa e não permitem elementos repetidos.
# Cenário: Filtrar os ativos únicos operados em um dia.
transacoes = {"PETR4", "VALE3", "PETR4", "ITUB4", "VALE3"}
print(f"Ativos únicos (Set - remove repetidos): {transacoes}")
```

```python
# 6. DICTS (Dicionários)

# Mapeamento de dados. Busca direta através de uma chave textual.
# Cenário: Painel de cotações com busca instantânea.
quadro_cotacoes = {
    "PETR4": 36.50,
    "VALE3": 62.10,
    "BBAS3": 27.80
}
# Acessando o valor informando a chave (nome do ativo)
print(f"A cotação de VALE3 é R$ {quadro_cotacoes['VALE3']}")
```

```python
# 7. OUTRAS APLICAÇÕES: INDEXAÇÃO E FATIAMENTO 

historico_precos = (34.50, 35.20, 36.00, 35.80, 36.50)

# Índice negativo [-1]: Acessa de trás para frente (pega o último elemento)
print(f"Último preço registrado t[-1]: {historico_precos[-1]}")

# Slicing [início:fim]: Extrai um intervalo (o índice de fim fica de fora)
print(f"Preços entre o índice 1 e 3 (t[1:4]): {historico_precos[1:4]}")
```

```python
# 8. OUTRAS APLICAÇÕES: CONCATENAÇÃO
# --------------------------------------------------------------
# Combina estruturas. Não altera as tuplas originais, cria uma terceira.
trimestre_1 = (1, 2)
trimestre_2 = (4, 5)

# Unindo os elementos
semestre = trimestre_1 + trimestre_2
print(f"Tuplas concatenadas (1,2) + (4,5) = {semestre}")
```

```python
# 9. FUNÇÕES QUE RETORNAM TUPLAS (def)

def avaliar_estudante(n1, n2):
    """
    Calcula a média de duas notas.
    Retorna uma tupla contendo o valor da média e a situação do aluno em texto.
    """
    media = (n1 + n2) / 2
    
    if media >= 7:
        situacao = "aprovado"
    else:
        situacao = "reprovado"
        
    # Ao retornar dois valores separados por vírgula, 
    # o Python cria e retorna implicitamente uma tupla!
    return media, situacao

# Capturando a tupla completa gerada pela função
resultado = avaliar_estudante(8.0, 6.0)
print(f"Resultado bruto da função (Tupla): {resultado}")

# Desempacotando o retorno da função direto em variáveis
media_final, status = avaliar_estudante(8.0, 6.0)
print(f"O aluno obteve média {media_final} e está {status}.")

```
